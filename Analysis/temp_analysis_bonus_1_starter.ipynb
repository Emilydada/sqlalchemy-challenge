### Compare June and December data across all years 

```python
from scipy import stats
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
Measurement = Base.classes.measurement
Station= Base.classes.station
session = Session(engine)
```

```python
# Filter data for desired months
sel = [Measurement.date, 
       Measurement.tobs]
data06= session.query(*sel).\
    filter(func.strftime("%m", Measurement.date) == "06").group_by (Measurement.date).all()
data12=session.query(*sel).\
    filter(func.strftime("%m", Measurement.date) == "12").group_by (Measurement.date).all()
display(data06,data12)
```

```python
# Identify the average temperature for June
df=pd.DataFrame(data06)
june_average = df.mean()
june_average
```

```python
# Identify the average temperature for December
df1=pd.DataFrame(data12)
dec_average = df1.mean()
dec_average
```

```python
# Create collections of temperature data
display(june_average,dec_average)
```

```python
# Run paired t-test


stats.shapiro(df[1])
stats.shapiro(df1[1])
```

```python
stats.levene(df[1],df1[1])
```

### Analysis

```python
res = stats.ttest_ind(df[1], df1[1], 
                      equal_var=True)

display(res)
```
